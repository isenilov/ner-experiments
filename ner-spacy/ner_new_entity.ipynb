# NER for new entity label

Training of the extractor to recognize new entity type.

```python
from util import load_conll
from os.path import join

train_data = load_conll(join("..", "data", "train.txt"))
test_data = load_conll(join("..", "data", "test.txt"))
valid_data = load_conll(join("..", "data", "valid.txt"))
print(train_data[10])
```

```python
new_labels = set()
for entry in train_data:
    for label in entry[1]["entities"]:
        new_labels.add(label[2])
new_labels
```

```python
import spacy
import random

nlp = spacy.blank('en')
ner = nlp.create_pipe('ner')
nlp.add_pipe(ner)

# """Set up the pipeline and entity recognizer, and train the new entity."""
# model = None

# if model is not None:
#     nlp = spacy.load(model)  # load existing spaCy model
#     print("Loaded model '%s'" % model)
# else:
#     nlp = spacy.blank('en')  # create blank Language class

# # Add entity recognizer to model if it's not in the pipeline
# # nlp.create_pipe works for built-ins that are registered with spaCy
# if 'ner' not in nlp.pipe_names:
#     ner = nlp.create_pipe('ner')
#     nlp.add_pipe(ner)
# # otherwise, get it, so we can add labels to it
# else:
#     ner = nlp.get_pipe('ner')

for label in new_labels:
    ner.add_label(label)   # add new entity label to entity recognizer
```

```python
%%time
from spacy.util import minibatch

#optimizer = nlp.begin_training()
n_iter = 400
for itn in range(n_iter):
    # i = 0
    random.shuffle(train_data)
    losses = {}
    batches = minibatch(train_data, 100)
    for batch in batches:
        texts, annotations = zip(*batch)
        nlp.update(texts, annotations, sgd=optimizer, drop=0.5, losses=losses)
        # if i%100 == 0: 
        #     print(losses)
        # i += 1
    print("Loss after epoch", itn, "=", losses["ner"])
```

```python
from spacy.gold import GoldParse
from spacy.scorer import Scorer

scorer = Scorer()
for input_, annot in test_data:
    doc_gold_text = nlp.make_doc(input_)
    gold = GoldParse(doc_gold_text, entities=annot["entities"])
    pred_value = nlp(input_)
    scorer.score(pred_value, gold)
print(scorer.scores)
```

```python
# save model to output directory
from datetime import datetime
from pathlib import Path
dt = datetime.now().strftime("%d-%m-%Y.%H-%M")


output_dir = Path("out")
if not output_dir.exists():
    output_dir.mkdir()
nlp.meta['name'] = "NER " + dt # rename model
nlp.to_disk(output_dir)
print("Saved model to", output_dir)

# test the saved model
print("Loading from", output_dir)
nlp2 = spacy.load(output_dir)
doc2 = nlp2("Testing the Spacy model, written in Python in Innopolis")
for ent in doc2.ents:
    print(ent.label_, ent.text)
```
